```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from packages import dataset, missing,visualization
devider = "=" * 50

while True:
    print("Choose an option:")
    print("1. Load Dataset File")
    print("2. Explore Data")
    print("3. Missing Values Handling")
    print("4. Dublicate Rows Removal")
    print("5. Calculate Borrowing Stats (NumPy)")
    print("6. Data Visualization (Matplotlib)")
    print("7. Statistical Analysis (Pandas)")
    print("8. Export Cleaned Dataset")
    print("9. Exit")

    choose = input("Enter your choice : ")
    match choose:
        case "1":
            print("\n========|| Dataset ||========\n")
            data = dataset.load_dataset()
            print(devider)
        case "2":
            print("\n========|| Explore Data ||========\n")
            dataset.explore_data(data)
            print(devider)
            
        case "3":
            print("\n========|| Missing Values Handling ||========\n")
            missing.missing_values_handling(data)
            print(devider)
        case "4":
            print("\n========|| Duplicate Rows Removal ||========\n")
            missing.duplicate_rows_removal(data)
            print(devider)
            
        case "5":
            print("\n========|| Calculate Borrowing Stats (NumPy) ||========\n")
            dataset.calculate_borrowing_stats(data)
            print(devider)
        
        case "6":
            print("\n========|| Data Visualization (Matplotlib) ||========\n")
            print("Select Visualization Type:")
            print("1. Bar Chart")
            print("2. Line Graph")
            print("3. Pie Chart")
            print("4. Heatmap")
            choose_viz = input("Enter your choice: ")
            if choose_viz == "1":
                visualization.plot_bar_chart(data)
            elif choose_viz == "2":
                visualization.plot_line_graph(data)
            elif choose_viz == "3":
                visualization.plot_pie_chart(data)
            elif choose_viz == "4":
                visualization.plot_heatmap(data)
            else:
                print("Invalid choice for visualization type.")
            print(devider)
            
        case "7":
            print("\n========|| Statistical Analysis (Pandas) ||========\n")
            missing.satatistical_analysis(data)
            print(devider)
        case "8":
            print("\n========|| Export Cleaned Dataset ||========\n")
            dataset.save_dataset(data)
            print(devider)
            
        case "9":
            print("Exiting the program. Goodbye!")
            break

        case _:
            print("Invalid choice. Please select a valid option.")
```
