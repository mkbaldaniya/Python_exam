```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from Packages import dataset, missing, visualization
print("\n========== > Data Analysis & Visualization Program < ==========")

devider = "-"*40

while True:
    print("\nPlease select an option:")
    print("1. Load Dataset")
    print("2. Explore Data")
    print("3. Perform DataFrame Operation")
    print("4. Handle Missing Data")
    print("5. Generate Descriptive Statistics")
    print("6. Data Visualization")
    print("7. Save Visualization")
    print("8. Exit")
    choice = int(input ("Enter Your Choice : "))
    
    match choice:
        case 1:
            print("\n========|| Dataset ||========\n")
            data = dataset.load_dataset()
            print(devider)
        case 2:
            print("\n========|| Explore Data ||========\n")
            print("1. Display the first 5 rows")
            print("2. Display the last 5 rows")
            print("3. Display column names")
            print("4. Display data types")
            print("5. Display basic info")
            choose = int(input("Choose an menu options : "))
            if choose == 1:
                print("\n~~~~~~~~~~~~~| First 5 Rows |~~~~~~~~~~~~~\n")
                print(data.head(5))
                print(devider)
            if choose == 2:
                print("\n~~~~~~~~~~~~~| Last 5 Rows |~~~~~~~~~~~~~\n")
                print(data.tail(5))
                print(devider)
            if choose == 3:
                print("\n~~~~~~~~~~~~~| Column Name |~~~~~~~~~~~~~\n")
                df_cols = data.columns
                mylist = []
                for i in df_cols:
                    # print(i)
                    mylist.append(i)
                print(mylist)
                print(devider)
            if choose == 4:
                print("\n~~~~~~~~~~~~~| Data Types |~~~~~~~~~~~~~\n")
                print(data.dtypes)
                print(devider)
            if choose == 5:
                print("\n~~~~~~~~~~~~~| Basic info |~~~~~~~~~~~~~\n")
                print(data.info())
                print(devider)
            else:
                print("Invalid option. Please select a valid option from the menu.")
        case 3:
            print("\n========|| DataFrame Operation ||========\n")
            print("1. Sort by Salary")
            print("2. Filter by Department")   
            choose = int(input("Choose an menu options : "))
            if choose == 1:
                print("\n~~~~~~~~~~~~~| Sort by Salary |~~~~~~~~~~~~~\n")
                sorted_data = data.sort_values(by="Salary", ascending=False)
                print(sorted_data)
                print(devider)
            if choose == 2:
                print("\n~~~~~~~~~~~~~| Filter by Department |~~~~~~~~~~~~~\n")
                department = input("Enter the department to filter by (e.g., 'Sales','Marketing','Finance'): ")
                filtered_data = data[data["Department"] == department]
                print(filtered_data)
                print(devider)
            else:
                print("Invalid option. Please select a valid option from the menu.")
        case 4:
            print("\n========|| Handle Missing Data ||========\n")
            print("1. Display rows with missing values")
            print("2. Fill missing values with mean")
            print("3. Drop rows with missing values")
            print("4. Replace missing values with a specific value")
            choose1 = int(input("Choose an menu options : "))
            if choose1 == 1:
                missing.display_missing_values(data)
                print(devider) 
            if choose1 == 2:
                missing.fill_missing_with_mean(data)
                print(devider)
            if choose1 == 3:
                data = missing.drop_rows_with_missing_values(data)
                print(devider)
            if choose1 == 4:
                value = input("Enter the value to replace missing values with: ")
                data = missing.replace_missing_with_value(data, value)
                print(devider)
            else:
                print("Invalid option. Please select a valid option from the menu.")
        case 5:
            print("\n========|| Descriptive Statistics ||========\n")
            print(data.describe())
            print(devider)
        case 6:
            print("\n========|| Data Visualization ||========\n")
            print("1. Bar Plot")
            print("2. Line Plot")
            print("3. Scatter Plot")
            print("4. Pie Chart")
            print("5. Histrogram")
            print("6. Stack Plot")
            choose = int(input("Choose an menu options : "))
            if choose == 1:
                visualization.bar_plot(data)
            if choose == 2:
                visualization.line_plot(data)
            if choose == 3:
                x_column = input("Enter the column name for x-axis: ")
                y_column = input("Enter the column name for y-axis: ")
                visualization.scatter_plot(data, x_column, y_column)
            if choose == 4:
                column = input("Enter the column name for pie chart: ")
                visualization.pie_chart(data, column)
            if choose == 5:
                visualization.histogram(data)
            if choose == 6:
                visualization.stack_plot(data)
            else:
                print("Invalid option. Please select a valid option from the menu.")
        case 7:
            print("\n========|| Save Visualization ||========\n")
            filename = input("Enter the filename to save the current plot (e.g., 'plot.png'): ")
            plt.savefig(filename)
        case 8:
            print("Exiting The program. Goodbye!")
            break
        case _:
            print("\nInvalid option. Please select a valid option from the menu.")
```
